```python
books_df =df.pivot_table(index='User.ID',
                                 columns='Book.Title',
                                 values='Book.Rating',aggfunc='mean', fill_value = 0)
```

```python
books_df.fillna(0,inplace=True)
```

```python
books_df
```

```python
distance=pairwise_distances(books_df,metric='cosine')
```

```python
distance
```

```python
similarity=1-pairwise_distances(books_df,metric='cosine')
similarity
```

```python
sim=pd.DataFrame(similarity)
```

```python
np.fill_diagonal(sim.values,0)
```
