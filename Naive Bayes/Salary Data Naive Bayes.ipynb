```python
from sklearn.preprocessing import StandardScaler, OrdinalEncoder,LabelEncoder
oe = OrdinalEncoder()
df.select_dtypes(object).columns
obj_enc=df.select_dtypes(object).columns
obj_enc
```

```python
obj_enc=obj_enc.drop('Salary')
obj_enc
```

```python
df[obj_enc]=oe.fit_transform(df[obj_enc])
df
```

```python
le= LabelEncoder()
df["Salary"]=le.fit_transform(df["Salary"])
df
```

```python
from sklearn.ensemble import IsolationForest
iso=IsolationForest(random_state=10,contamination="auto")
out=iso.fit_predict(df)
out
```

```python
df["outliers"]=out
df["outliers"].value_counts()
```

```python
# Dropping The ouliers
df=df[df["outliers"]== 1]
df
```

```python
df
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python

from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.30, random_state=42)
xtrain.shape, xtest.shape, ytrain.shape, ytest.shape
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
xtrain=scaler.fit_transform(xtrain)
xtest=scaler.transform(xtest)
```

```python
xtrain
```

```python
xtest
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
```

```python
nb = MultinomialNB()
nb.fit(xtrain,ytrain)
ypred = nb.predict(xtest)
print(classification_report(ytest,ypred))
print(nb.score(xtrain,ytrain))
print(nb.score(xtest,ytest))
```
